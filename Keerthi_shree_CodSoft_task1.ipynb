```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

X_train_imputed = imputer.fit_transform(X_train)
X_train = X_train.dropna()
Y_train = Y_train[X_train.index]

```

```python
data_titanic["Age"].fillna(0,inplace=True)
data_titanic["Fare"].fillna(0,inplace=True)

data_titanic["Age"]=data_titanic["Age"].astype(int)
data_titanic["Fare"]=data_titanic["Fare"].astype(int)

data_titanic.dropna(subset=["Age","Fare"],inplace=True)

data_titanic["Age"]=data_titanic["Age"].astype(int)
data_titanic["Fare"]=data_titanic["Fare"].astype(int)

data_titanic["Age"]=data_titanic["Age"].astype(float)
data_titanic["Fare"]=data_titanic["Fare"].astype(float)


```

```python
X = data_titanic.drop(columns=['PassengerId', 'Survived'], axis=1)
Y = data_titanic['Survived']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)

training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy_score_of_training_data:', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)

test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy_score_of_test_data:', test_data_accuracy)
```
